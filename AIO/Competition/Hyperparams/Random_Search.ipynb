Random Search sẽ lấy mẫu search space một cách ngẫu nhiên. Hiệu quả trên những data với high-dimensional space và thường được sử dụng rộng rãi trong thực tế. Nhược điểm của Random Search là không sử dụng thông tin từ những lần thử nghiệm trước đó để chọn ra những config parameter tiếp theo. Với Random Search việc tìm ra best hyperparameters phụ thuộc vào yếu tố "may mắn".

Random Search hoạt động cực kỳ hiệu quả và rất dễ hiểu. Mặc dù những lựa chọn là ngẫu nhiên tuy nhiên những sự ngẫu nhiên này không phải là một **naive selection**. Thực tế nó hoạt động như một cách lấy mẫu ngẫu nhiên trong thống kê: điểm chính của Random Search là chúng ta thực hiện đủ random tests, chúng ta sẽ có nhiều khả năng tìm các thông số phù hợp mà không lãng phí quá nhiều tài nguyên.

Một số hệ thống **AutoML** sẽ dựa vào Random Search khi có quá nhiều tham số để thiết lập và thử nghiệm.

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
X, y = make_classification(n_samples=300, n_features=50,
                          n_informative=10,
                          n_redundant=25, n_repeated=15,
                          n_clusters_per_class=5,
                          flip_y=0.05, class_sep=0.5,
                          random_state=0)
```

```python
from sklearn.utils.fixes import loguniform
from sklearn import svm
from sklearn import model_selection

svc = svm.SVC()
svc = svm.SVC(probability=True, random_state=1)

search_dict = {'kernel': ['linear', 'rbf'],
                'C': loguniform(1, 1000),
                'gamma': loguniform(0.0001, 0.1)
              }

scorer = 'accuracy'
search_func = model_selection.RandomizedSearchCV(estimator=svc,param_distributions=search_dict, n_iter=10,
                                                  scoring=scorer, n_jobs=-1, cv=5) 
```

```python
# Training
search_func.fit(X, y)
print (search_func.best_params_)
print (search_func.best_score_)
```
