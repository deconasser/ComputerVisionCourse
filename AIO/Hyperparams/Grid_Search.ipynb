Grid Search là một phương thức cho phép tìm kiếm các hyperparameter một cách hiệu quả nhất. Tuy nhiên Grid Search sẽ **không phù hợp** với **high-dimensional space**. Bởi vỗi mọi parameter, chúng ta cần set một tập giá trị muốn test. Và sau đó sẽ thực hiện training model với tổ hợp của các parameter đó. Đây là lý do vì sao nói Grid Search là hiệu quả nhất bởi nó cho phép chúng ta thử nghiệm mọi trường hợp (vét cạn). Đây là một thuật toán đơn giản, hiệu quả - tuy nhiên thời gian training sẽ rất lâu. Để tối ưu Grid Search yêu cầu chúng ta phải hiểu rõ Machine Learning Algorithms - xác định được khoảng hiệu quả và vai trò của mỗi parameter. 

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
X, y = make_classification(n_samples=300, n_features=50,
                          n_informative=10,
                          n_redundant=25, n_repeated=15,
                          n_clusters_per_class=5,
                          flip_y=0.05, class_sep=0.5,
                          random_state=0)
```

Sau khi tạo một bộ dữ liệu classification thì chúng ta sẽ sử dụng SVM để tạo một model phân lớp hiệu quả cùng với một tập search space cho Grid Search. Khi đó **kernel function** của SVC sẽ quyết định đến các bộ hyperparameter khác nhau. 

```python
from sklearn import svm
from sklearn import model_selection
from sklearn.utils.fixes import loguniform

svc = svm.SVC()
svc = svm.SVC(probability=True, random_state=1)

search_grid = [
  {'C': [1, 10, 100, 1000], 'kernel': ['linear']},
  {'C': [1, 10, 100, 1000], 'gamma': [0.001, 0.0001],
  'kernel': ['rbf']}
]

scorer = 'accuracy'
```

```python
search_func = model_selection.GridSearchCV(estimator=svc,
                                          param_grid=search_grid,
                                          scoring=scorer,
                                          n_jobs=-1,
                                          cv=5)
```

```python
# Training 
search_func.fit(X, y)

print(search_func.best_params_)
print(search_func.best_score_)
```
